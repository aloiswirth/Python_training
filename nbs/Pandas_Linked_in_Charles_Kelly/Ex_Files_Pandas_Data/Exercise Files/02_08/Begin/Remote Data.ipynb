```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from pandas_datareader import data, wb
```

### Yahoo finance

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2021, 7, 15)
yahoo_df = data.DataReader("F", 'yahoo', start, end)
yahoo_df.plot()
```

### FRED
source: http://quant-econ.net/py/pandas.html

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2021, 7, 15)
unrate_df = data.DataReader('UNRATE', 'fred', start, end)
unrate_df.plot()
```

### Google

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2021, 7, 15)
google_df = data.DataReader("F", 'microsoft', start, end)
google_df.plot()
```
